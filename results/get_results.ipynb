```python
import pandas as pd  
import os  

# 指定根文件夹路径  
root_folder = '/data/home/lichengzhang/zhoujun/Rehaimian/logs/MLPModel_2'  # 替换为实际路径  

# 创建一个空的列表来存储 DataFrame  
dataframes = []  

# 遍历根文件夹下的所有子文件夹  
for folder in sorted(os.listdir(root_folder)):  
    folder_path = os.path.join(root_folder, folder)  

    # 检查当前路径是否是文件夹  
    if os.path.isdir(folder_path):  
        # 构建 predictions.csv 的完整路径  
        csv_file_path = os.path.join(folder_path, 'predictions.csv')  
        
        # 检查文件是否存在  
        if os.path.isfile(csv_file_path):  
            # 读取 CSV 文件并添加到 DataFrame 列表  
            df = pd.read_csv(csv_file_path)  
            dataframes.append(df)  

# 合并所有 DataFrame  
combined_df = pd.concat(dataframes, ignore_index=True)  

# 输出合并后的 DataFrame  
combined_df["Time"] = pd.to_datetime(combined_df['Time'], format='ISO8601').dt.tz_convert('Asia/Shanghai')

```

```python
combined_df
```

```python
y_tag = pd.read_feather("/data/home/lichengzhang/zhoujun/HaimianData/20250325/y_tag_458_300750.SZ.ftr")
```

```python
result_df = pd.merge(combined_df, y_tag, on="Time", how="left")
result_df = result_df[result_df['y_action'] == 1]
result_df.describe()
```
